```python

import tensorflow as tf
from tensorflow.keras import layers, regularizers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import DenseNet169
from tensorflow.keras.applications.densenet import preprocess_input
from tensorflow.keras.optimizers import AdamW
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

from tensorflow.keras import mixed_precision
mixed_precision.set_global_policy("mixed_float16")

# ==============================================================
# Paths (Validation folder MUST exist separately)
# ==============================================================
TRAIN_PATH = "archive/brisc2025/classification_task/train"
VAL_PATH   = "archive/brisc2025/classification_task/val"   

# ==============================================================
# Data Augmentation — Strong for real medical images
# ==============================================================
train_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input,
    rotation_range=30,
    width_shift_range=0.20,
    height_shift_range=0.20,
    zoom_range=0.25,
    brightness_range=[0.6, 1.4],
    horizontal_flip=True,
    vertical_flip=True,
    fill_mode='nearest'
)

val_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)

train_gen = train_datagen.flow_from_directory(
    TRAIN_PATH,
    target_size=(256, 256),
    batch_size=32,   
    class_mode='categorical',
    shuffle=True
)

val_gen = val_datagen.flow_from_directory(
    VAL_PATH,
    target_size=(256, 256),
    batch_size=32,
    class_mode='categorical',
    shuffle=False
)

# ==============================================================
# DenseNet169 Base
# ==============================================================
base = DenseNet169(
    weights="imagenet",
    include_top=False,
    input_shape=(256, 256, 3)
)

# Stage 1 — Freeze all layers
for layer in base.layers:
    layer.trainable = False

# ==============================================================
# Classification Head — Upgraded for Better Accuracy
# ==============================================================
x = layers.GlobalAveragePooling2D()(base.output)
x = layers.Dense(512, activation='relu', kernel_regularizer=regularizers.l2(0.001))(x)
x = layers.Dropout(0.45)(x)
x = layers.BatchNormalization()(x)
x = layers.Dense(256, activation='relu')(x)
x = layers.Dropout(0.35)(x)

output = layers.Dense(4, activation='softmax', dtype="float32")(x)

model = tf.keras.Model(inputs=base.input, outputs=output)

# ==============================================================
# Stage 1 Training — Top Layers Only
# ==============================================================
optimizer1 = AdamW(learning_rate=3e-4, weight_decay=1e-4)

model.compile(
    optimizer=optimizer1,
    loss='categorical_crossentropy',
    metrics=['accuracy', tf.keras.metrics.AUC(name='auc')]
)

history1 = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=20,
    callbacks=[
        EarlyStopping(monitor='val_auc', patience=5, restore_best_weights=True),
        ReduceLROnPlateau(monitor='val_auc', factor=0.45, patience=3)
    ]
)

# ==============================================================
# Stage 2 — Fine-Tune Final DenseNet Layers
# ==============================================================
for layer in base.layers[-120:]:      # Training deeper layers for accuracy gain
    layer.trainable = True

optimizer2 = AdamW(learning_rate=1e-5, weight_decay=1e-5)

model.compile(
    optimizer=optimizer2,
    loss='categorical_crossentropy',
    metrics=['accuracy', tf.keras.metrics.AUC(name='auc')]
)

history2 = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=25,
    callbacks=[
        EarlyStopping(monitor='val_auc', patience=6, restore_best_weights=True),
        ReduceLROnPlateau(monitor='val_auc', factor=0.45, patience=3)
    ]
)

# ==============================================================
# Save Final Model
# ==============================================================
model.save("DenseNet169_BRISC256_Accuracy.h5")
print("\n Model saved as DenseNet169_BRISC256_Accuracy.h5\n")
```
