```python
import tensorflow as tf
from tensorflow.keras import layers, models
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
IMG_SIZE = 256
IMAGE_PATH = "image.jpg"
MASK_PATH = "mask.png"
EPOCHS = 200

img = cv2.imread(IMAGE_PATH)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
img = img /255.0

mask = cv2.imread(MASK_PATH, cv2.IMREAD_GRAYSCALE)
mask = cv2.resize(mask, (IMG_SIZE, IMG_SIZE))
mask = mask / 255.0
mask = (mask > 127).astype("float32")
mask = np.expand_dims(mask, axis=-1)

X = np.expand_dims(img, axis=0)
Y = np.expand_dims(mask, axis=0)

def conv_block(x, filters):
    x = layers.Conv2D(filters, 3, padding="same", activation="relu")(x)
    x = layers.Conv2D(filters, 3, padding="same", activation="relu")(x)
    return x

def encoder_block(x, filters):
    c = conv_block(x, filters)
    p = layers.MaxPooling2D()(c)
    return c, p

def decoder_block(x, skip, filters):
    x = layers.Conv2DTranspose(filters, 2, strides=2, padding="same")(x)
    x = layers.Concatenate()([x, skip])
    x = conv_block(x, filters)
    return x

inputs = layers.Input((IMG_SIZE, IMG_SIZE, 3))

c1, p1 = encoder_block(inputs, 64)
c2, p2 = encoder_block(p1, 128)
c3, p3 = encoder_block(p2, 256)

bottleneck = conv_block(p3, 512)

d1 = decoder_block(bottleneck, c3, 256)
d2 = decoder_block(d1, c2, 128)
d3 = decoder_block(d2, c1, 64)

outputs = layers.Conv2D(1, 1,  activation="sigmoid")(d3)

model = models.Model(inputs, outputs)

model.compile(
    optimizer="adam",
    loss="binary_crossentropy",
    metrics=["accuracy"]
)

model.summary()
history = model.fit(X, Y, epochs=EPOCHS)
```

```python
pred = model.predict(X)[0, :, :, 0]

binary_pred = (pred > 0.5).astype("uint8") * 255
cv2.imwrite("predicted_mask_binary.png", binary_pred)

plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.title("Image")
plt.imshow(img)
plt.axis("off")

plt.subplot(1,3,2)
plt.title("Ground Truth Mask")
plt.imshow(mask.squeeze(), cmap="gray")
plt.axis("off")

plt.subplot(1,3,3)
plt.title("Predicted Mask")
plt.imshow(pred, cmap="gray")
plt.axis("off")

plt.show()
```

```python
import os

OUTPUT_DIR = "outputs"
os.makedirs(OUTPUT_DIR, exist_ok=True)

pred_path = os.path.join(OUTPUT_DIR, "predicted_mask.png")

cv2.imwrite(
    pred_path,
    (pred * 255).astype("uint8")
)

print("Saved prediction to:", pred_path)
```
